# Lab 1 — Sandbox Executor

**Goal:** Run arbitrary Python code safely inside a subprocess. Capture stdout, stderr, exit code, and elapsed time.

This is the foundation of every code agent. Before we can fix code we need to *run* it.

```python
# Setup — make sure we can import our sandbox module
import sys, os
sys.path.insert(0, os.path.dirname(os.path.abspath('__file__')))
from sandbox import run_code, run_pytest
print('Sandbox loaded ✓')
```

## 1. Running simple code

```python
result = run_code("print('Hello from the sandbox!')")
print(result.summary())
```

## 2. Capturing errors

```python
bad_code = """
x = [1, 2, 3]
print(x[10])  # IndexError
"""
result = run_code(bad_code)
print(result.summary())
print('\nexited with code:', result.exit_code)
```

## 3. Timeout protection
An infinite loop should be killed, not hang your notebook.

```python
infinite_loop = "while True: pass"
result = run_code(infinite_loop, timeout=2.0)
print(result.summary())
print('Timed out?', result.timed_out)
```

## 4. Running pytest
Write a solution and tests, run pytest, inspect results.

```python
source = """
def add(a, b):
    return a - b  # BUG: should be +
"""

tests = """
def test_add():
    assert add(2, 3) == 5

def test_add_negative():
    assert add(-1, 1) == 0
"""

result = run_pytest(tests, source)
print(result.summary())
```

## 5. Fixing the bug and re-running

```python
fixed_source = """
def add(a, b):
    return a + b  # FIXED
"""

result = run_pytest(tests, fixed_source)
print(result.summary())
```
