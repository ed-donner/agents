```python
def read_linkedin_info():
    reader = PdfReader("./me/ryan_john.pdf")
    linkedin = ""
    for page in reader.pages:
        text = page.extract_text()
        if text:
            linkedin += text
    return linkedin
```

```python
def review_summary():
    summary = ""
    with open("./me/ryan_john_summary.txt", "r", encoding="utf-8") as f:
        summary = f.read()
    return summary
```
