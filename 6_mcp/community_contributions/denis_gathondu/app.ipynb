# MCP Variation of the job application

##### I am using SSE for my MCP since I'm using an older PC that only has WSL 1. To run this notebook, you have to install `mcp-proxy` via `uv add mcp-proxy`.

##### inside the folder run `uv run applicant_server.py` to start my mcp server.

##### inside the folder run `mcp-proxy --port 8000 -- npx -y markdown2pdf-mcp@latest` to run the markdown to pdf mcp server

##### I'm using rapid api, make sure to have rapid api key set in your .env. You can get it here https://rapidapi.com/

##### Also set `RAPID_API_HOST="linkedin-job-search-api.p.rapidapi.com"` You have to subscribe to the api here https://rapidapi.com/fantastic-jobs-fantastic-jobs-default/api/linkedin-job-search-api

##### To send emails, I'm using smtplib so you need to set `FROM_EMAIL`, `TO_EMAIL`, `GOOGLE_APP_PASSWORD`, `EMAIL_HOST`, `EMAIL_PORT`


```python
from agents import Runner, trace
from agents.mcp import MCPServerSse
from applicant_agents import ApplicantAgents
from IPython.display import Markdown, display

params = {"url": "http://127.0.0.1:8000/sse"}
pdf_params = {"url": "http://127.0.0.1:8001/sse"}
async with (
    MCPServerSse(params, client_session_timeout_seconds=60) as app_server,
    MCPServerSse(pdf_params, client_session_timeout_seconds=60) as pdf_server,
):
    agents = ApplicantAgents("Denis Gathondu")
    manager = await agents.create_agent(app_server, pdf_server)
    with trace("Application Manager"):
        result = await Runner.run(manager, "Run the job application pipeline.")
    display(Markdown(result.final_output))
```
