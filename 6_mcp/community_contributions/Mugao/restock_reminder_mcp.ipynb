simulation
This demo shows how MCP transforms communication from:

"People trying to remember things"

into:

"Systems coordinating automatically"

```python
inventory = {
    "niacinamide serum": 0,
    "vitamin c serum": 5,
    "retinol cream": 2
}

restock_requests = {}

print("Initial Inventory:", inventory)
```

```python
def check_inventory(product):
    product = product.lower()
    stock = inventory.get(product)

    if stock is None:
        return "Product not found"
    elif stock == 0:
        return "Out of stock"
    return f"In stock: {stock}"



# print(check_inventory("niacinamide serum"))
# print(check_inventory("vitamin c serum"))
```

```python
def request_restock_alert(product, customer):
    product = product.lower()

    if product not in restock_requests:
        restock_requests[product] = []

    restock_requests[product].append(customer)

    return f"{customer} will be notified when {product} is restocked."


# print(request_restock_alert("niacinamide serum", "Mugao"))
# print("Current Requests:", restock_requests)
```

restock product and test

```python
def restock_product(product, quantity):
    product = product.lower()
    inventory[product] = inventory.get(product, 0) + quantity
    return f"{product} restocked. New quantity: {inventory[product]}"


# print(restock_product("niacinamide serum", 10))
# print("Updated Inventory:", inventory)
```

notify customers

```python
def notify_customers(product):
    product = product.lower()
    customers = restock_requests.get(product, [])

    if not customers:
        return "No customers to notify."

    messages = []
    for customer in customers:
        messages.append(f"Notification sent to {customer}: {product} is now available!")

    return "\n".join(messages)


# print(notify_customers("niacinamide serum"))
```

analyze demand

```python
def analyze_demand():
    if not restock_requests:
        return "No demand data."

    results = []
    for product, users in restock_requests.items():
        results.append(f"{product}: {len(users)} interested customers")

    return "\n".join(results)


# print(analyze_demand())
```
