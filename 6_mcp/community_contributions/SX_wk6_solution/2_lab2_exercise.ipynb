```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
from agents.mcp import MCPServerStdio
from IPython.display import display, Markdown

load_dotenv(override=True)
```

### MCP Server

```python
params = {"command": "uv", "args": ["run", "today_server.py"]}
async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as server:
    mcp_tools = await server.list_tools()
```

```python
mcp_tools
```

```python
instructions = "You are able to use tool to get today's date, and answer questions about today's date."
request = "What's today's date?"
model = "gpt-4o-mini"
```

```python

async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as mcp_server:
    agent = Agent(name="assistant", instructions=instructions, model=model, mcp_servers=[mcp_server])
    with trace("assistant"):
        result = await Runner.run(agent, request)
    display(Markdown(result.final_output))
```

### MCP Client

```python
from today_client import get_today_tools_openai, list_today_tools
mcp_tools = await list_today_tools()
print(mcp_tools)
openai_tools = await get_today_tools_openai()
print(openai_tools)
```

```python
request = "What's today's date?"

with trace("today_client"):
    agent = Agent(name="assistant", instructions=instructions, model=model, tools=openai_tools)
    result = await Runner.run(agent, request)
    display(Markdown(result.final_output))
```
