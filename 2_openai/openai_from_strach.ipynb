```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
```

```python
# 1. Manually clear any Azure "ghost" variables in memory
for var in ["OPENAI_API_TYPE", "OPENAI_API_BASE", "AZURE_OPENAI_ENDPOINT", "OPENAI_API_VERSION"]:
    if var in os.environ:
        del os.environ[var]
```

```python

# 1. Load your .env file
load_dotenv()

# 2. Initialize the client
# This automatically looks for 'OPENAI_API_KEY' in your .env
client = OpenAI()

try:
    # 3. Make a simple request
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": "Say 'The API is working!' in a creative way."}
        ]
    )
    
    # 4. Print the result
    print("--- Response ---")
    print(response.choices[0].message.content)
    
except Exception as e:
    print(f"--- Error ---")
    print(f"It might still be activating. Error: {e}")
```
