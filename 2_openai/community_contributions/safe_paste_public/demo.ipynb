## Safe public paste — multi-agent demo

Orchestration: **Scanner → Redaction coach** (high/critical) **or Explainer** (low/medium, with `input_guardrail`).

Set `OPENAI_API_KEY` (e.g. via repo `.env`). **Run with the kernel working directory = this folder** (`safe_paste_public`), or change `ROOT` below.

```python
from pathlib import Path
import sys

from dotenv import load_dotenv

ROOT = Path.cwd()
assert (ROOT / "orchestrator.py").exists(), (
    "Set the notebook cwd to safe_paste_public (folder containing orchestrator.py)."
)
sys.path.insert(0, str(ROOT))
load_dotenv(override=True)

from orchestrator import SafePasteManager
```

```python
# Pick a sample: clean_traceback | leaky_log | medium_pii
SAMPLE = ROOT / "sample_inputs" / "leaky_log.txt"
text = SAMPLE.read_text(encoding="utf-8")
print(SAMPLE.name, "—", len(text), "chars")
```

```python
async def show():
    mgr = SafePasteManager()
    async for chunk in mgr.run(text):
        print(chunk)


await show()
```
