```python
# The imports

from agents import Agent, OpenAIProvider, Runner, RunConfig, trace

```

```python
# Load .env from cwd or parent (agents/.env), then configure DeepSeek (OpenAI-compatible API)
import os
from pathlib import Path

from dotenv import load_dotenv
from agents import OpenAIProvider, RunConfig

for env_path in (
    Path.cwd() / ".env",
    Path.cwd().parent / ".env",
    Path.cwd().parent.parent / ".env",
):
    if env_path.is_file():
        load_dotenv(env_path, override=True)
        break

deepseek_key = os.getenv("DEEPSEEK_API_KEY")
if not deepseek_key:
    raise RuntimeError(
        "DEEPSEEK_API_KEY is not set. Add DEEPSEEK_API_KEY=... to your agents/.env "
        "(https://platform.deepseek.com/api_keys)."
    )

# Chat Completions only — DeepSeek does not support OpenAI's Responses API
deepseek_provider = OpenAIProvider(
    api_key=deepseek_key,
    base_url="https://api.deepseek.com/v1",
    use_responses=False,
)
deepseek_run_config = RunConfig(model_provider=deepseek_provider)
```

```python

# Make an agent with name, instructions, model (DeepSeek model id)

agent = Agent(name="Jokester", instructions="You are a joke teller", model="deepseek-chat")
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(
        agent,
        "Tell a joke about Autonomous AI Agents",
        run_config=deepseek_run_config,
    )
    print(result.final_output)
```
