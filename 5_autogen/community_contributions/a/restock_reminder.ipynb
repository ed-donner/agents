```python
import os
from dotenv import load_dotenv
from openai import OpenAI
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
load_dotenv()
```

basic agent

```python
class RetailAgent:
    def __init__(self, name):
        self.name = name

    def respond(self, message):
        response = client.chat.completions.create(
            model="gpt-4.1",
            messages=[
                {"role": "system", "content": "You are a helpful retail assistant."},
                {"role": "user", "content": message}
            ]
        )
        return response.choices[0].message.content
```

```python
agent = RetailAgent("RetailAssistant")

print(agent.respond("Do you have niacinamide serum?"))
```

dummy inventory system

```python
inventory = {
    "niacinamide serum": 0,
    "vitamin c serum": 5,
    "retinol cream": 2
}
```

```python
def check_inventory(product):
    product = product.lower()
    stock = inventory.get(product)

    if stock is None:
        return "Product not found"
    elif stock == 0:
        return "Out of stock"
    return f"In stock: {stock}"
```

```python
print("Test 1:", check_inventory("niacinamide serum"))
print("Test 3:", check_inventory("unknown product"))
```

```python
class SmartRetailAgent:
    def handle_request(self, product, customer_name):
        stock_status = check_inventory(product)

        if stock_status == "Out of stock":
            return f"{product} is out of stock. Would you like to be notified?"
        
        return f"{product} is available. {stock_status}"
```

```python
agent = SmartRetailAgent()

print("Test 1:", agent.handle_request("niacinamide serum", "Mugao"))
print("Test 2:", agent.handle_request("vitamin c serum", "Mugao"))
```

```python
restock_requests = {}

def request_notification(product, customer):
    if product not in restock_requests:
        restock_requests[product] = []

    restock_requests[product].append(customer)
    return f"{customer} will be notified when {product} is restocked."
```

```python
print(request_notification("niacinamide serum", "Mugao"))
print(request_notification("niacinamide serum", "AlSx"))

print("Current requests:", restock_requests)
```

output

```python
agent = SmartRetailAgent()

product = "niacinamide serum"

# Step 1: Check product
response = agent.handle_request(product, "Mugao")
print("Step 1:", response)

# User subscribes
subscription = request_notification(product, "Mugao")
print("Step 2:", subscription)

# Simulate restock
inventory[product] = 10
print("Step 3: Product restocked")

# Notify users
print("Step 4: Sending notifications...")
for user in restock_requests.get(product, []):
    print(f"Notification sent to {user}: {product} is now available!")
```

can we use the same data to check which products are most in demand?

```python
def analyze_demand():
    if not restock_requests:
        print("No demand data yet.")
        return

    for product, users in restock_requests.items():
        print(f"{product}: {len(users)} interested customers")


analyze_demand()
```
