```python
from pydantic import BaseModel
```

```python
class Evaluate(BaseModel):
    is_acceptable: bool
    feedback: str
```

```python
gemini_key = os.getenv('GOOGLE_API_KEY')
gemini = OpenAI(api_key=gemini_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
model_name = "gemini-2.5-flash"
```

```python
evaluator_system_prompt = f"You are an evaluator that decides whether a response to a question is acceptable. \
You are provided with a conversation between a User and an Agent. Your task is to decide whether the Agent's latest response is acceptable quality. \
The Agent is playing the role of {name} and is representing {name} on their website. \
The Agent has been instructed to be professional and engaging, as if talking to a potential client or future employer who came across the website. \
The Agent has been provided with context on {name} in the form of their summary and LinkedIn details. Here's the information:"

evaluator_system_prompt += f"\n\n## Summary:\n{summary}\n\n## LinkedIn Profile:\n{cv}\n\n"
evaluator_system_prompt += f"With this context, please evaluate the latest response, replying with whether the response is acceptable and your feedback."
```

```python
def evaluator_user_prompt(query, reply, history):
    user_prompt = f"Here's the conversation between the User and the Agent: \n\n{history}\n\n"
    user_prompt += f"Here's the latest message from the User: \n\n{query}\n\n"
    user_prompt += f"Here's the latest response from the Agent: \n\n{reply}\n\n"
    user_prompt += "Please evaluate the response, replying with whether it is acceptable and your feedback."
    return user_prompt
```

```python
def evaluate_response(query, response, history):
    message = [{'role': 'system', 'content': evaluator_system_prompt}] + [{'role': 'user', 'content': evaluator_user_prompt(query, response, history)}]
    response = gemini.chat.completions.parse(model=model_name, messages=message, response_format=Evaluate)
    return response.choices[0].message.parsed
```

```python
def rerun_prompt(query, response, history, feedback):
    updated_system_prompt = system_prompt + "\n\n## Previous answer rejected\nYou just tried to reply, but the quality control rejected your reply\n"
    updated_system_prompt += f"## Your attempted answer:\n{response}\n\n"
    updated_system_prompt += f"## Reason for rejection:\n{feedback}\n\n"
    messages = [{"role": "system", "content": updated_system_prompt}] + history + [{"role": "user", "content": query}]
    response = openai.chat.completions.create(model="gpt-4o-mini", messages=messages)
    return response.choices[0].message.content
```

```python
def chat(query, history):
    message = [{'role': 'system', 'content': system_prompt + " Answer it in pig latin"}] + history + [{'role': 'user', 'content': query}]
    response = openai.chat.completions.create(model='gpt-4o-mini', messages=message)
    response = response.choices[0].message.content
    attempts = 0
    while True:
        evaluate = evaluate_response(query, response, history)
        print(evaluate)
        if evaluate.is_acceptable is False:
            attempts += 1
            if attempts <= 3:
                print("Rerunning . . .")
                response = rerun_prompt(query, response, history, feedback=evaluate.feedback)
            else:
                return "Max attempts reached. Not giving correct output."
        else:
            return response
```

```python
record_user_info = {
    'name': 'record_user_details',
    'description': 'Always use this tool to store user info, if they provide their email and all.',
    'parameters': {
        'type': 'object',
        'properties': {
            'email': {
                'type': 'string',
                'description': 'Whenever user enters his or her email, store it'
            },
            'address': {
                'type': 'string',
                'description': 'Whenever user enters his or her address'
            }
        },
        'required': ['email']
    },
}

record_unkown_question = {
    'name': 'record_unknown_question',
    'description': 'Always use this tool to stroe when user asks and unkown quesition or irrelevent, which you couldn\'t answer',
    'parameters': {
        'type': 'object',
            'properties': {
            'query': {
                'type': 'string',
                'description': 'The question that couldn\'t be answered'
            }
            },
        'required': ['question']
    },
}
```

```python
tools = [{'type': 'function', 'function': record_user_info}, {'type': 'function', 'function': record_unkown_question}]
```

```python
def handle_tool_calls(tools):
    results = []
    for tool in tools:
        name = tool.function.name
        arguments = tool.function.arguments
        print(f"Tool called: {name}")
        print(f"Stored: {arguments}")
        results.append({'role': 'tool', 'content': arguments, 'tool_call_id': tool.id})
        return results
```

```python
def chat(message, history):
    message = [{'role': 'system', 'content': system_prompt}] + history + [{'role': 'user', 'content': message}]
    done = False
    while not done:
        response = openai.chat.completions.create(model='gpt-4.1-mini', messages=message, tools=tools)
        msg = response.choices[0].message
        if response.choices[0].finish_reason == 'tool_calls':
            results = handle_tool_calls(response.choices[0].message.tool_calls)
            message.append(msg)
            message.extend(results)
            print(message)
        else:
            done = True
    return response.choices[0].message.content
```
