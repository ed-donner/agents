```python
from dotenv import load_dotenv
from openai import OpenAI
import os
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
```

```python
openai = OpenAI()
client = OpenAI(api_key=os.environ.get('DEEPSEEK_API_KEY'), base_url="https://api.deepseek.com")
```

<h2 style="color:#ff7800;">Exercise</h2>
Using deepseek

```python
# First create the messages:
question1="Pick a business area that might be worth exploring for an agentic AI opportunity"
messages = [
            {"role": "system", "content": "You are a helpful assistant"},
            {"role": "user", "content": question1}
        ]

# Then make the first call:
response1 = client.chat.completions.create(
    model="deepseek-chat",
# Then read the business idea:
    messages=messages,
    stream=False)

business_idea = response1.choices[0].message.content
display(Markdown(business_idea))


# And repeat! In the next message, include the business idea within the message
```

```python
question2="present a pain-point in Commercial Real Estate (CRE) Brokerage & Asset Management industry \
            - something challenging that might be ripe for an Agentic solution."
messages = [
            {"role": "system", "content": question1+'\n'+business_idea},
            {"role": "user", "content": question2}
        ]
response2 = client.chat.completions.create(
    model="deepseek-chat",
# Then read the business idea:
    messages=messages,
    stream=False)

pain_point = response2.choices[0].message.content
display(Markdown(pain_point))
```

```python
question3="Propose the Agentic solution."
messages = [
            {"role": "system", "content": question1+'\n'+business_idea+'\n'+question2+'\n'+pain_point},
            {"role": "user", "content": question3}
        ]
response3 = client.chat.completions.create(
    model="deepseek-chat",
# Then read the business idea:
    messages=messages,
    stream=False)

agentic_solution = response3.choices[0].message.content
display(Markdown(agentic_solution))
```

```python
Ollama_client = OpenAI(
    base_url='http://localhost:11434/v1/',
    api_key='ollama',  # Required but ignored by Ollama
)
```

```python
# First create the messages:
question1="Pick a business area that might be worth exploring for an agentic AI opportunity"
messages = [
            {"role": "system", "content": """You are an Agentic Reasoning Engine. 
        Before providing a final answer, you must:
        1. Analyze the user's intent.
        2. Break the problem into logical sub-tasks.
        3. Evaluate potential constraints.
        4. Output your 'Reasoning' section clearly before the 'Final Answer'."""},
            {"role": "user", "content": question1}
        ]

# Then make the first call:
response1 = Ollama_client.chat.completions.create(
    model="gemma4:e2b",
# Then read the business idea:
    messages=messages,
    stream=False)

business_idea = response1.choices[0].message.content
display(Markdown(business_idea))
# And repeat! In the next message, include the business idea within the message
```

```python
question2="present a pain-point in Autonomous Knowledge & Workflow Management for Mid-Market Operations (The Operations Agent). \
            - something challenging that might be ripe for an Agentic solution."
messages = [
            {"role": "system", "content": """You are an Agentic Reasoning Engine. 
        Before providing a final answer, you must:
        1. Analyze the user's intent.
        2. Break the problem into logical sub-tasks.
        3. Evaluate potential constraints.
        4. Output your 'Reasoning' section clearly before the 'Final Answer'."""+'\n'+question1+'\n'+business_idea},
            {"role": "user", "content": question2}
        ]
response2  = Ollama_client.chat.completions.create(
    model="gemma4:e2b",
# Then read the business idea:
    messages=messages,
    stream=False)

pain_point = response2.choices[0].message.content
display(Markdown(pain_point))
```
