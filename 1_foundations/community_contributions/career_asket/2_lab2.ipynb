```python
# Anthropic: use native API if ANTHROPIC_API_KEY is set, else OpenRouter from .env

model_name = "claude-sonnet-4-5"

if anthropic_api_key:
    claude = Anthropic()
    response = claude.messages.create(model=model_name, messages=messages, max_tokens=1000)
    answer = response.content[0].text
elif openrouter_api_key:
    client = OpenAI(api_key=openrouter_api_key, base_url="https://openrouter.ai/api/v1")
    # Use stable OpenRouter model IDs; try fallback if one returns 500
    for or_model in ["anthropic/claude-3.5-sonnet", "anthropic/claude-3-haiku"]:
        try:
            response = client.chat.completions.create(
                model=or_model, messages=messages, max_tokens=1000, timeout=120.0
            )
            answer = response.choices[0].message.content
            break
        except Exception as e:
            if "500" in str(e) or "InternalServerError" in type(e).__name__:
                continue
            raise
    else:
        answer = "(OpenRouter returned 500 for Claude models)"
else:
    answer = "(Anthropic: set ANTHROPIC_API_KEY or OPENROUTER_API_KEY in .env)"

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# It's nice to know how to use "zip"
try:
    competitors
    answers
except NameError:
    competitors = []
    answers = []
for competitor, answer in zip(competitors, answers):
    print(f"Competitor: {competitor}\n\n{answer}")
```

```python
# Let's bring this together - note the use of "enumerate"

try:
    answers
except NameError:
    answers = []
together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
try:
    question
except NameError:
    question = "(Run the cells above to set the question.)"
try:
    together
except NameError:
    together = "(No responses yet; run the provider cells above.)"
try:
    competitors
except NameError:
    competitors = []

judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
# Judgement time!

try:
    OpenAI
except NameError:
    from openai import OpenAI
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=judge_messages,
    timeout=120.0,
)
results = response.choices[0].message.content
print(results)
```

```python
# OK let's turn this into results!

try:
    json
except NameError:
    import json

try:
    results
except NameError:
    results = '{"results": []}'
try:
    competitors
except NameError:
    competitors = []

results_dict = json.loads(results)
ranks = results_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
if not ranks:
    print("(No rankings yet; run the judge cell above first.)")
```

### Which pattern(s) did this lab use?

1. **Multi-model / parallel agents** – The same question was sent to multiple LLM providers (OpenAI, Anthropic, Gemini, DeepSeek, Groq, Ollama). Each acts as a "competitor" answering in parallel.

2. **Judge / evaluator** – A separate LLM call takes all responses and ranks them (best to worst), acting as a judge.

### Adding another pattern: **Summarizer**

Below we add a **Summarizer** pattern: after ranking, we ask an LLM to summarize the **winning** response in a few bullet points (e.g. for a busy mayor). This is a common agentic pattern to turn long outputs into actionable summaries.

```python
# --- New pattern: Summarizer — summarize the winning response for the "mayor" ---
try:
    results_dict
except NameError:
    results_dict = {"results": []}
try:
    competitors
    answers
except NameError:
    competitors = []
    answers = []

if results_dict.get("results") and competitors and answers:
    winner_rank_1 = results_dict["results"][0]  # best competitor number (1-based)
    winner_name = competitors[int(winner_rank_1) - 1]
    winner_answer = answers[int(winner_rank_1) - 1]

    summarizer_prompt = f"""The following is the winning response from model "{winner_name}" to a complex policy question. Summarize it in 3–5 short bullet points suitable for a busy decision-maker. Be precise and preserve key numbers and recommendations.

Response to summarize:
---
{winner_answer[:6000]}
---
Bullet-point summary:"""

    try:
        openai
    except NameError:
        from openai import OpenAI
        openai = OpenAI()

    summary_response = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[{"role": "user", "content": summarizer_prompt}],
        timeout=60.0,
    )
    summary = summary_response.choices[0].message.content
    print("Summary of the winning response (for the mayor):")
    print(summary)
else:
    print("Run the cells above to get competitors, answers, and judge results first.")
```

### Where this pattern shows up in practice

- **Customer support & FAQs** – Send the same question to several models (or the same model with different prompts), then use a judge or validator to pick the best or safest answer before showing it to users.
- **Legal, compliance, contracts** – Generate or compare clauses with multiple models; a judge or rules can select the most appropriate or flag risky wording.
- **Content moderation & policy** – Multiple models label content; aggregate or judge their outputs to improve consistency and reduce bias from a single model.
- **Summarization & reporting** – E.g. summarise the “winning” long answer (as in the Summarizer step) for executives or dashboards.
- **R&D and internal tools** – Compare model behaviour on the same inputs to choose a provider, tune prompts, or monitor quality.

### Why it matters for business

- **Accuracy and safety** – One model can hallucinate or be inconsistent; comparing and judging reduces the chance of shipping a bad answer.
- **Audit and explainability** – You keep multiple candidates and a ranking or score, which helps with debugging and compliance.
- **Vendor and model strategy** – You can switch or combine providers (OpenAI, Anthropic, open-source, etc.) without changing the overall design.

### Trade-offs to keep in mind

- **Cost** – N models + 1 judge means N+1 API calls per task; use when quality justifies the extra cost.
- **Latency** – Running models in parallel keeps delay down; the judge adds one more round-trip.
- **Complexity** – You need clear prompts, a stable judge criterion, and (often) parsing/validation of the judge output.
