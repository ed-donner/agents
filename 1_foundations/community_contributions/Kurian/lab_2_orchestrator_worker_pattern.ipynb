## Exercise answer: Orchestrator–Workers pattern

Extension of Lab 2 using Anthropic's [**Orchestrator–Workers**](https://www.anthropic.com/engineering/building-effective-agents) workflow.

### Roles

| Role | Model | Responsibility |
|------|-------|----------------|
| **Orchestrator** | GPT (`gpt-5-mini`) | Break the hard question into sub-questions and assign each to a worker |
| **Workers** | GPT, Claude, Gemini, DeepSeek, Groq | Answer only the sub-question assigned to them |
| **Aggregator** | GPT (`gpt-5-mini`) — same as orchestrator | Combine worker outputs into one final answer |

### Worker pool

| Worker | Provider | Model |
|--------|----------|-------|
| GPT | OpenAI | `gpt-5.4-mini` |
| Claude | Anthropic | `claude-sonnet-4-5` |
| Gemini | Google | `gemini-2.5-flash` |
| DeepSeek | DeepSeek | `deepseek-chat` |
| Groq | Groq | `openai/gpt-oss-120b` |

### Flow

```
Original question
       │
       ▼
┌──────────────────┐
│  GPT Orchestrator │  decompose + assign workers
└────────┬─────────┘
         │
    ┌────┴────┬────────┬──────────┬────────┐
    ▼         ▼        ▼          ▼        ▼
  GPT     Claude   Gemini   DeepSeek   Groq
    │         │        │          │        │
    └────┬────┴────────┴──────────┴────────┘
         ▼
┌──────────────────┐
│  GPT Aggregator   │  synthesize → unified answer
└──────────────────┘
```

### Patterns in this lab

**Original section (above):** Parallelization (voting) — the same question goes to every model; a judge ranks the answers.

**This extension:** Orchestrator–Workers — GPT dynamically splits the question into different subtasks, delegates each to a specialized worker, then GPT aggregates the partial answers into one response.

The key difference from parallelization: workers do **different** subtasks, not the same task repeated.

````python
import json 

worker_types = ['gpt', 'claude', 'gemeni', 'deepseek', 'groc']
answers = []
user_question = "A runaway trolley is headed toward five people tied to the track, and you can pull a lever to divert it onto another track where one person is tied—do you pull the lever or do nothing?"


# Orchestrator 
orchestrator = OpenAI() 
messages = [     
{
        "role": "system", 
        "content":  f""" You are an LLM orchestrator that will take the user's input queston pass it to the list of 
        workers shown here worker_types: {worker_types}. You will pass it to each worker in order. 
        Make sure you give the worker agents enough details for you to answer the question
        You must iterate the list of workers shown above. You must STRICTLY respond in json as follows (ALWAYS and no special characters): 
            - agent: <next_worker> from the list. 
            - message: message to pass each worker.
        If all agents are iterated, then you must reply WORKFLOW_FINISHED as  agent: null and message: WORKFLOW_FINISHED in the json format. (Strictly iterate through all the workers).
        Strictly avoid visiting the same agent twice.
        When returning json, be mindful that it must accomodate json.loads(...) so please no special characters in json response or spaces.
        Ask the agents to answer to explain in layman's terms without heavy academic terms. 
        Strictly follow worder -> gpt -> claude -> gemeni -> deepseek -> groc
        """
}, 
{
        "role": "user", 
        "content": user_question   

}]

# Initialize workers 
gpt_worker = OpenAI()
gpt_model = "gpt-5.4-mini"

claude_worker = Anthropic()
claude_model = "claude-sonnet-4-5"

gemeni_worker =  OpenAI(api_key=google_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
gemeni_model = "gemini-2.5-flash"


deepseek_worker =  OpenAI(api_key=deepseek_api_key, base_url="https://api.deepseek.com/v1")
deepseek_model = "deepseek-chat"


groc_worker = OpenAI(api_key=groq_api_key, base_url="https://api.groq.com/openai/v1")
groc_model = "openai/gpt-oss-120b"


response_json = [{'agent': None, 'message': ''}]

visited = []

while not "WORKFLOW_FINISHED" in response_json[0]['message']:
        response = orchestrator.chat.completions.create(model="gpt-5.4-mini", messages=messages)
        response_value = "["+response.choices[0].message.content.strip()+"]"
        print(response_value)
        response_json = json.loads(response_value)
        agent = response_json[0]["agent"] 

        if agent == "gpt": 
                # Call GPT to get response 
                worker_response = gpt_worker.chat.completions.create(
                        model=gpt_model,
                        messages=[{'role': 'user', 'content': response_json[0]['message']}]
                )

                worker_response_answer = worker_response.choices[0].message.content 
                answers.append({'agent': agent, 'answer': worker_response_answer})
                messages.append({'role': 'user', "content": f"GPT answered : {worker_response_answer}"})
        
        if agent == "claude": 
                # Call claude response 
                worker_response = claude_worker.messages.create(model=claude_model, messages=[{'role': 'user', 'content': response_json[0]['message']}], max_tokens=1000)
                worker_response_answer = worker_response.content[0].text
                answers.append({'agent': agent, 'answer': worker_response_answer})
                messages.append({'role': 'user', "content": f"Claude answered : {worker_response_answer}"})


        if agent == "gemeni": 
                # Call gemeni response 
                worker_response = gemeni_worker.chat.completions.create(model=gemeni_model, messages=[{'role': 'user', 'content': response_json[0]['message']}])
                worker_response_answer = worker_response.choices[0].message.content
                answers.append({'agent': agent, 'answer': worker_response_answer})
                messages.append({'role': 'user', "content": f"Gemeni answered : {worker_response_answer}"})


        if agent == "deepseek": 
                # Call deepseek response 
                worker_response = deepseek_worker.chat.completions.create(model=deepseek_model, messages=[{'role': 'user', 'content': response_json[0]['message']}])
                worker_response_answer = worker_response.choices[0].message.content
                answers.append({'agent': agent, 'answer': worker_response_answer})
                messages.append({'role': 'user', "content": f"Deepseek answered : {worker_response_answer}"})

        
        if agent == "groc": 
                # Call grok 
                worker_response = groc_worker.chat.completions.create(model=groc_model, messages=[{'role': 'user', 'content': response_json[0]['message']}])
                worker_response_answer = worker_response.choices[0].message.content
                answers.append({'agent': agent, 'answer': worker_response_answer})
                messages.append({'role': 'user', "content": f"Groc answered : {worker_response_answer}"})


print("Answer Output")
for item in answers: 
        print(item, "\n")



judge_model = OpenAI()
gpt_model = "gpt-5.4-mini"

answers_prompt = [{
        "role": "system", 
        "content": """ 
        Here are the list of answers from several different agent and you will rank each agent's answer based on ethicalness, logical thinking and reasoning. 
        Strictly Return output as json as follows: 
                {
                        "ethicalness": ["gpt": 5, claude: 4 .... ],  //list from descending order from left to right
                        "logical_thinking": ["gpt": 5, claude: 4. ...], //same here, list in descending order
                        "reasoning": ["gpt": 5, claude: 4 ...] //this is descending in order as well reasoning. 
                }


        """
}, 

{
        "role": "user", 
        "content": f""" Here are the answers from agents: {answers}"""

}]


judge_response = judge_model.chat.completions.create(model=gpt_model, messages=answers_prompt)

judge_response_text = judge_response.choices[0].message.content


print(judge_response_text)




````
