```python
open_ai_key = os.getenv('OPENAI_API_KEY')
```

```python
question = '''Give me a list of models that I can use to evaluate the intelligence of LLMs. 

response = openai.chat.completions.create(
    model="{gpt-4.1-mini}",
    messages=messages
)

I want the list of models that I can use in the above code. I just want the list of models, nothing else no codeblock or markdown formatting.

'''
messages = [{"role": "user", "content": question}]
```

```python
openai = OpenAI(api_key=open_ai_key)

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)
print(response.choices[0].message.content)
```

```python
models_to_use = []
for model in response.choices[0].message.content.split(","):
    models_to_use.append(model)
print(models_to_use)
```

```python
question = '''Give me a mediums leetcode question from any topic. It should have all the constraints and edge cases and have a sample input and output. I dont want any codeblock or markdown formatting.
'''
messages = [{"role": "user", "content": question}]

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)
leetcode_question = response.choices[0].message.content
```

```python
ordered_models = {}
for index,model in enumerate(models_to_use):
    print(f"{index+1}. {model}")
```

```python
new_question = (
    f"You are given a question. Please read it properly and answer accordingly. "
    f"The question is: '{leetcode_question}'. "
    "Now I want you to give me 2 solutions: "
    "1) brute force solution and 2) optimal solution. "
    "You need to give reasons why the optimal solution is optimal and "
    "why the brute force solution is not optimal. Moreover, you need to give "
    "the time and space complexity of both solutions."
)

messages = [{"role": "user", "content": new_question}]
```

```python
new_question
```

```python
answers = []
model_used = []
for model in models_to_use:

    # print(f"Model: '{model.strip()}'")
    try:
        response = openai.chat.completions.create(
            model=f'{model.strip()}',
            messages=messages
        )
        answers.append(response.choices[0].message.content)
        model_used.append(model.strip())
    except Exception as e:
        print(f"Error: {e}")
        # answers.append(f"Error: {e}")

print(answers)

```

```python
# Let's bring this together - note the use of "enumerate"

together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
together
```

```python
judge = f"""You are judging a competition between {len(model_used)} competitors.
Each model has been given this question:

{leetcode_question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
```

```python
messages = [{"role": "user", "content": judge}]

openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=messages,
)

results = response.choices[0].message.content
```

```python
print(results)
```

```python
ranking = json.loads(results)
for index,value in enumerate(ranking['results']):
    print(f'Rank {index+1} goes to Mode:{model_used[int(value)-1]}')
```
