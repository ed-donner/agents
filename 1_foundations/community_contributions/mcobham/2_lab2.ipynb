```python
# OpenAI client libraries with the right base_url and key
# If this surprises you, please see Guide 9 in the Guides folder!

anthropic = OpenAI(api_key=anthropic_api_key, base_url=ANTHROPIC_BASE_URL)
deepseek = OpenAI(api_key=deepseek_api_key, base_url=DEEPSEEK_BASE_URL)
gemini = OpenAI(api_key=google_api_key, base_url=GEMINI_BASE_URL)
groq = OpenAI(api_key=groq_api_key, base_url=GROQ_BASE_URL)
grok = OpenAI(api_key=grok_api_key, base_url=GROK_BASE_URL)
openrouter = OpenAI(api_key=openrouter_api_key, base_url=OPENROUTER_BASE_URL)
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

Prompt Chaining - This process starts with us prompting for a question and then posing that question to AI, to then ask AI to judge the answers based on the question. 
Routing - The question is sent out to multiple AI agents, in this case to judge different models, returing their input.
Orchestrator-Worker - The overall structure is Parallelization organised by AI.
Evaluator-Optimizer - The LLM output is validated by another.

```python
#AI means Animal Identifier

import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)

openai_api_key = os.getenv('OPENAI_API_KEY')

request = """
You are the orchestrator designing a short intelligence evaluation challenge for AI models. 

Create a single, self-contained challenge instruction that will later be given independently to three different AI models. The challenge must require each model to:

- Secretly choose one real animal (any vertebrate or invertebrate; encourage some variety in size, habitat, or taxonomic group, but avoid extremely obscure species that almost no one would know).
- Write a detailed description of that animal covering appearance, behavior, habitat, diet, unique adaptations, sounds or movement, and any other distinctive traits.
- Never name the animal, its common name, scientific name, or use any obvious synonym or giveaway phrase that would immediately reveal its identity.
- Keep the entire description between 40 and 70 words. No more, no less. Prefer short, information-dense sentences.

Output ONLY the finished challenge instruction text that the three worker models will receive. Do not choose any animals yourself, do not give examples of animals, and do not add any extra commentary.
"""
messages = [{"role": "user", "content": request}]

openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-5.5",
    messages=messages,
)
question = response.choices[0].message.content
print(question)
```

```python
judge = f"""
You are an expert naturalist and careful reasoner. Below are three independent descriptions of animals. Each description was written by a different AI that was forbidden from naming the animal.

For each description:
1. Reason step-by-step: list the most diagnostic traits, eliminate unlikely candidates, and narrow the possibilities.
2. State your final identification (common name + brief justification).
3. Score the description on three axes (1–5 each):
   - Diagnostic power (how uniquely the details point to one animal)
   - Avoidance of leakage (did it accidentally give the name away?)
   - Richness & accuracy of knowledge

At the end, briefly rank the three descriptions from most to least effective at enabling correct identification. 

{together}
"""
messages = [{"role": "user", "content": judge}]

openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-5.5",
    messages=messages,
)
response = response.choices[0].message.content
display(Markdown(response))
```
