## Routing pattern: choose the right model for the task

This version rewrites the lab to use Anthropic's **Routing** workflow pattern.

Instead of sending the same question to every model, a router LLM classifies the question and chooses one specialist route. The workflow then calls only the selected downstream model. Finally, Grok via OpenRouter audits whether the routing decision made sense.

```python
# OpenAI Compatible URLs

ANTHROPIC_BASE_URL = "https://api.anthropic.com/v1/"
DEEPSEEK_BASE_URL = "https://api.deepseek.com/v1"
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
GROQ_BASE_URL = "https://api.groq.com/openai/v1"
GROK_BASE_URL = "https://api.x.ai/v1"
OPENROUTER_BASE_URL = "https://openrouter.ai/api/v1"
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
# OpenAI client libraries with the right base_url and key
# If this surprises you, please see Guide 9 in the Guides folder!

anthropic = OpenAI(api_key=anthropic_api_key, base_url=ANTHROPIC_BASE_URL)
deepseek = OpenAI(api_key=deepseek_api_key, base_url=DEEPSEEK_BASE_URL)
gemini = OpenAI(api_key=google_api_key, base_url=GEMINI_BASE_URL)
groq = OpenAI(api_key=groq_api_key, base_url=GROQ_BASE_URL)
grok = OpenAI(api_key=grok_api_key, base_url=GROK_BASE_URL)
openrouter = OpenAI(api_key=openrouter_api_key, base_url=OPENROUTER_BASE_URL)
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
competitors = []
answers = []
route_decision = {}
messages = [{"role": "user", "content": question}]
```

```python
def record(model_name, answer):
    competitors.append(model_name)
    answers.append(answer)
    display(Markdown(answer))
```

```python
# Define the available specialist routes.
# Each route points to one model/client pair and describes when the router should choose it.

routes = {
    "general_reasoning": {
        "description": "General nuanced reasoning, balanced judgment, and concise synthesis.",
        "client": openai,
        "model": "gpt-5.4-nano",
        "api_key": openai_api_key,
    },
    "ethical_social_reasoning": {
        "description": "Questions involving values, tradeoffs, social judgment, ambiguity, or human consequences.",
        "client": anthropic,
        "model": "claude-sonnet-4-6",
        "api_key": anthropic_api_key,
    },
    "fast_structured_answer": {
        "description": "Questions that need a direct, structured, low-latency answer.",
        "client": gemini,
        "model": "gemini-3.1-flash-lite",
        "api_key": google_api_key,
    },
    "technical_analysis": {
        "description": "Technical, analytical, code-adjacent, or systems reasoning questions.",
        "client": deepseek,
        "model": "deepseek-v4-flash",
        "api_key": deepseek_api_key,
    },
    "creative_alternative": {
        "description": "Creative, lateral, unconventional, or style-sensitive questions.",
        "client": openrouter,
        "model": "moonshotai/kimi-k2.6",
        "api_key": openrouter_api_key,
    },
    "local_private": {
        "description": "Local-only route for experiments where privacy or offline execution matters more than quality.",
        "client": ollama,
        "model": "llama3.2:1b",
        "api_key": "ollama",
        "local": True,
    },
}

available_routes = {
    route_key: route
    for route_key, route in routes.items()
    if route.get("api_key") or route.get("local")
}

route_options = "\n".join(
    f"- {route_key}: {route['description']} Use model {route['model']}."
    for route_key, route in available_routes.items()
)

print(route_options)
```

```python
# The router classifies the question and chooses one downstream route.

router_prompt = f"""You are a router for an agentic workflow.
Your job is to choose the single best route for the user's question.

Question:
{question}

Available routes:
{route_options}

Respond with JSON only, using this exact shape:
{{"route_key": "one of the route keys", "reason": "brief reason for this choice"}}
"""

router_messages = [{"role": "user", "content": router_prompt}]
router_response = openai.chat.completions.create(
    model="gpt-5.4-nano",
    messages=router_messages,
    reasoning_effort="none",
)
route_decision_text = router_response.choices[0].message.content
print(route_decision_text)
```

```python
route_decision = json.loads(route_decision_text)
selected_route_key = route_decision.get("route_key")

if selected_route_key not in available_routes:
    selected_route_key = "general_reasoning"
    route_decision = {
        "route_key": selected_route_key,
        "reason": "Router returned an unavailable route, so the workflow fell back to general reasoning."
    }

selected_route = available_routes[selected_route_key]
selected_model = selected_route["model"]

print(f"Selected route: {selected_route_key}")
print(f"Selected model: {selected_model}")
print(f"Reason: {route_decision.get('reason')}")
```

```python
# Now run only the selected specialist model.

response = selected_route["client"].chat.completions.create(
    model=selected_model,
    messages=messages,
)
answer = response.choices[0].message.content

record(selected_model, answer)
```

```python
# Optional audit: use Grok via OpenRouter to check whether the routing choice was sensible.

routing_audit = f"""You are auditing an agentic routing workflow.

Original question:
{question}

Available routes:
{route_options}

Chosen route: {selected_route_key}
Chosen model: {selected_model}
Router reason: {route_decision.get('reason')}

Answer produced by the selected model:
{answer}

Evaluate whether the selected route was appropriate. Respond with JSON only:
{{"routing_was_appropriate": true or false, "better_route_if_any": "route key or none", "reason": "brief explanation"}}
"""

audit_messages = [{"role": "user", "content": routing_audit}]
print(routing_audit)
```

```python
# Judgement time!
# Grok is "The most truth-seeking large language model in the world."

model_name = "x-ai/grok-4.3"

response = openrouter.chat.completions.create(model=model_name, messages=audit_messages)
audit_results = response.choices[0].message.content
print(audit_results)
```

```python
# OK let's turn this into routing audit results.

audit_results_dict = json.loads(audit_results)
print(f"Routing appropriate: {audit_results_dict['routing_was_appropriate']}")
print(f"Better route if any: {audit_results_dict['better_route_if_any']}")
print(f"Reason: {audit_results_dict['reason']}")
```
