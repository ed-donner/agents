# Lab 1 exercise: discovering an Agentic AI opportunity

This solution uses a three-step prompt chain. The first LLM call selects a promising business area, the second identifies a painful problem in that area, and the third proposes a concise Agentic AI solution. Each result is passed into the next prompt so the final proposal is grounded in the business area and pain point discovered earlier.

```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)
openai = OpenAI()

# Ask the first LLM call to select a business area.
messages = [{
    "role": "user",
    "content": "Pick one business area that could benefit from an Agentic AI opportunity. Respond with only the business area.",
}]

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages,
)

business_idea = response.choices[0].message.content
print("Business area:", business_idea)

# Ask the second LLM call to identify a pain point in that business area.
messages = [{
    "role": "user",
    "content": f"For the business area '{business_idea}', describe one painful business problem that could be solved by an Agentic AI system. Respond in 2-3 sentences.",
}]

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages,
)

pain_point = response.choices[0].message.content
print("\nPain point:", pain_point)

# Ask the third LLM call to propose an Agentic AI solution.
messages = [{
    "role": "user",
    "content": f"Business area: {business_idea}\nPain point: {pain_point}\n\nPropose a simple Agentic AI solution. Include the agent's goal, the tools it might use, and the business value. Keep it concise.",
}]

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages,
)

agentic_solution = response.choices[0].message.content
print("\nAgentic AI solution:\n", agentic_solution)
```
