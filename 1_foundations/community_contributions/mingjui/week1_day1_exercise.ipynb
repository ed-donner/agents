```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display

load_dotenv(override=True)
openai = OpenAI()

# Prompt 1
prompt_1 = "Please pick a business area that might be worth exploring for an Agentic AI opportunity. Respond with only the name of the business area."
messages_1 = [{"role": "user", "content": prompt_1}]
response_1 = openai.chat.completions.create(
    model="gpt-5.4-mini",
    messages=messages_1
)

business_area = response_1.choices[0].message.content
display(Markdown(f"**Selected Business Area:** {business_area}"))
print("\n" + "="*50 + "\n")

# Prompt 2
prompt_2 = f"Present a pain-point in {business_area} - something challenging that might be ripe for an Agentic solution"
messages_2 = [{"role": "user", "content": prompt_2}]
response_2 = openai.chat.completions.create(
    model="gpt-5.4-mini",
    messages=messages_2
)

pain_point = response_2.choices[0].message.content
display(Markdown(f"**Pain Point:**\n{pain_point}"))
print("\n" + "="*50 + "\n")

# Prompt 3
prompt_3 = f"""
In {business_area}, the pain point is:
{pain_point}

Please propose an Agentic AI solution. Detail how multiple agents or tools would collaborate to resolve this.
"""
messages_3 = [{"role": "user", "content": prompt_3}]
response_3 = openai.chat.completions.create(
    model="gpt-5.4-mini",
    messages=messages_3
)

agentic_solution = response_3.choices[0].message.content
display(Markdown(agentic_solution))
```
