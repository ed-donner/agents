```python
import os
from openai import OpenAI 
from IPython.display import Markdown, display
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

openai = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
messages = [{"role": "user", "content": "What is 2+2?"}]
```

```python
response = openai.chat.completions.create(
    model="llama3.2",
    messages=messages
)

print(response.choices[0].message.content)
```

```python
question = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]
```

```python
response = openai.chat.completions.create(
    model="llama3.2",
    messages=messages
)

question = response.choices[0].message.content

print(question)
```

```python
messages = [{"role": "user", "content": question}]
```

```python
response = openai.chat.completions.create(
    model="llama3.2",
    messages=messages
)

answer = response.choices[0].message.content
print(answer)
```

```python
display(Markdown(answer))
```

```python
# Exercise 1

messages = [{"role": "user", "content": "Pick a business idea that might be worthy exploring for an Agentic AI opportunity. Respond only with the business idea."}]

response = openai.chat.completions.create(
    model="llama3.2",
    messages=messages
)

business_idea = response.choices[0].message.content

print(business_idea)

messages = [{"role": "user", "content": f"Here is the business idea: {business_idea}. What is a pain-point in that industry - something challenging that might be ripe for an Agentic AI solution?"}]

response = openai.chat.completions.create(
    model="llama3.2",
    messages=messages
)

answer = response.choices[0].message.content

display(Markdown(answer))
```
