```python
def record_user_details(email, name="Name not provided", notes="not provided"):

    message = f"New contact interest from {name} | email: {email} | notes: {notes}"

    push(message)

    return {"status": "recorded"}
```

```python
def record_unknown_question(question):

    message = f"Unknown question asked: {question}"

    push(message)

    return {"status": "recorded"}
```

```python
record_user_details_json = {
    "name": "record_user_details",
    "description": "Use this tool when a user provides their email or wants to get in touch",
    "parameters": {
        "type": "object",
        "properties": {
            "email": {"type": "string"},
            "name": {"type": "string"},
            "notes": {"type": "string"}
        },
        "required": ["email"]
    }
}
```

```python
record_unknown_question_json = {
    "name": "record_unknown_question",
    "description": "Use this tool when the assistant cannot answer a question",
    "parameters": {
        "type": "object",
        "properties": {
            "question": {"type": "string"}
        },
        "required": ["question"]
    }
}
```

```python
tools = [
    {"type": "function", "function": record_user_details_json},
    {"type": "function", "function": record_unknown_question_json}
]
```

```python
def handle_tool_calls(tool_calls):

    results = []

    for tool_call in tool_calls:

        tool_name = tool_call.function.name
        arguments = json.loads(tool_call.function.arguments)

        print(f"Tool called: {tool_name}")

        tool = globals().get(tool_name)

        if tool:
            result = tool(**arguments)
        else:
            result = {}

        results.append({
            "role": "tool",
            "tool_call_id": tool_call.id,
            "content": json.dumps(result)
        })

    return results
```

```python
reader = PdfReader("me/linkedin.pdf")
linkedin = ""
for page in reader.pages:
    text = page.extract_text()
    if text:
        linkedin += text

with open("me/summary.txt", "r", encoding="utf-8") as f:
    summary = f.read()

name = "Ed Donner"
```

```python
#system prompt
system_prompt = f"""
You are acting as {name}.

You are answering questions on {name}'s personal website.

Your role is to represent {name} professionally when visitors ask about:

- career
- experience
- projects
- skills
- background

If someone asks about hiring, collaboration, or work opportunities,
respond positively and encourage them to get in touch.

If a user shares their email or wants to connect,
use the tool record_user_details.

If you cannot answer a question, use the tool record_unknown_question.

Be professional, friendly, and helpful.
"""

system_prompt += f"""

## SUMMARY
{summary}

## LINKEDIN PROFILE
{linkedin}

Stay in character as {name}.
"""
```

```python
def chat(message, history):
    messages = [{"role": "system", "content": system_prompt}] + history + [{"role": "user", "content": message}]
    done = False
    while not done:

        # This is the call to the LLM - see that we pass in the tools json

        response = openai.chat.completions.create(model="openai/gpt-4o-mini", messages=messages, tools=tools)

        finish_reason = response.choices[0].finish_reason
        
        # If the LLM wants to call a tool, we do that!
         
        if finish_reason=="tool_calls":
            message = response.choices[0].message
            tool_calls = message.tool_calls
            results = handle_tool_calls(tool_calls)
            messages.append(message)
            messages.extend(results)
        else:
            done = True
    return response.choices[0].message.content
```

```python
gr.ChatInterface(chat, type="messages").launch()
```
