```python
from dotenv import load_dotenv
from openai import OpenAI
import os

load_dotenv(override=True)


openai_api_key = os.getenv("OPENAI_API_KEY")


openai = OpenAI()
```

## Assignment

1. Ask LLM for a business area that is worth exploring for a Agentic AI opportunity
2. Ask AI to identify a pain-point in that industry - ripe for an agentic solution
3. Then the third LLM call to propose the solution.


```python
# First create the messages:

messages = [
    {
        "role": "user",
        "content": "Identify a business niches that is worth exploring for Agentic AI Opportunities.",
    }
]

# Then make the first call:

response = openai.chat.completions.create(model="gpt-5.4", messages=messages)

# Then read the business idea:

business_area = response.choices[0].message.content

print(business_area)

# And repeat! In the next message, include the business area within the message
```

```python
messages.extend(
    [
        {
            "role": "assistant",
            "content": business_area,
        },
        {
            "role": "user",
            "content": "Now identify 3 pain points within the business niche that is worth exploring for Agentic AI Opportunities, it should be challenging and ripe for Agentic AI solutions.",
        },
    ]
)

response = openai.chat.completions.create(model="gpt-5.4", messages=messages)

pain_points = response.choices[0].message.content

print(pain_points)
```

```python
messages.extend(
    [
        {
            "role": "assistant",
            "content": pain_points,
        },
        {
            "role": "user",
            "content": "Now propose a solution of each of the pain points mentioned above.",
        },
    ]
)

response = openai.chat.completions.create(model="gpt-5.4", messages=messages)

solutions = response.choices[0].message.content

print(solutions)
```
