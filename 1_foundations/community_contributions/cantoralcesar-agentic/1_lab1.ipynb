````python
# First let's do an import. If you get an Import Error, double check that your Kernel is correct..

from dotenv import load_dotenv
````

#### **¿Qué hace?**

`python-dotenv` es una librería que lee un archivo de texto llamado `.env` (generalmente ubicado en la raíz de tu proyecto) y carga las variables de entorno que contiene en el entorno de ejecución de tu programa.

#### **¿Por qué se usa?**

Es una práctica común para gestionar configuraciones sensibles (como claves API, contraseñas de bases de datos, tokens secretos) sin hardcodearlas en el código fuente. De este modo:

* Puedes compartir tu código públicamente sin exponer credenciales
* Cada desarrollador o entorno (desarrollo, producción, testing) puede tener sus propios valores
* El archivo `.env` se añade al `.gitignore` para no subirlo a repositorios

````python
# Next it's time to load the API keys into environment variables
# If this returns false, see the next cell!

load_dotenv(override=True)  # Carga el .env en las variables de entorno
````


#### **¿Qué hace `override=True`?**

Por defecto, `load_dotenv()` **no sobreescribe** variables de entorno que ya existen en el sistema. Es decir, si una variable ya está definida en tu entorno (por ejemplo, en tu terminal o en la configuración del servidor), el valor del archivo `.env` se **ignora**.

Con `override=True`, le dices a `load_dotenv` que **sí sobreescriba** las variables existentes con los valores del archivo `.env`.

#### **Ejemplo práctico**

Imagina que en tu terminal ya tienes:

```bash
export API_KEY="valor_del_sistema"
```

Y tu archivo `.env` tiene:

```bash
API_KEY="valor_del_archivo"
```

| Código | Resultado de `os.getenv("API_KEY")` |
|--------|-------------------------------------|
| `load_dotenv()` | `"valor_del_sistema"` (se mantiene el del sistema) |
| `load_dotenv(override=True)` | `"valor_del_archivo"` (se sobreescribe con el del `.env`) |

#### **¿Cuándo usar `override=True`?**

- **Desarrollo local**: Cuando quieres que el `.env` tenga prioridad sobre cualquier variable que pueda estar definida en tu sistema.
- **Testing**: Para asegurarte de que se usen exactamente los valores de tu archivo de configuración de pruebas.
- **Contenedores/Docker**: A veces es útil forzar que el `.env` prevalezca.

#### **¿Cuándo NO usarlo?**

- **Producción**: Si prefieres que las variables definidas en el servidor (más seguras) tengan prioridad sobre un archivo `.env` que podría existir accidentalmente.

#### **Resumen**

| Parámetro | Comportamiento |
|-----------|----------------|
| `override=False` (por defecto) | El `.env` solo rellena variables **ausentes**. Las existentes no se tocan. |
| `override=True` | El `.env` **sobreescribe** todas las variables, existan o no. |

````python
# And now - the all important import statement
# If you get an import error - head over to troubleshooting in the Setup folder
# Even for other LLM providers like Gemini, you still use this OpenAI import - see Guide 9 for why

from openai import OpenAI
````

````python
# And now we'll create an instance of the OpenAI class
# If you're not sure what it means to create an instance of a class - head over to the guides folder (guide 6)!
# If you get a NameError - head over to the guides folder (guide 6)to learn about NameErrors - always instantly fixable
# If you're not using OpenAI, you just need to slightly modify this - precise instructions are in the AI APIs guide (guide 9)

openai_client = OpenAI()  
````



| Parte | Qué es |
|-------|--------|
| `openai` | Es el **paquete/librería** que instalas con `pip install openai`. Contiene varias clases y funciones, entre ellas OpenAI |
| `OpenAI` | Es la **clase** dentro del paquete. Es el “molde” que define cómo crear un cliente para conectarte a la API. |
| `openai_client` | Es el **nombre de la variable** donde guardas la instancia |
| `OpenAI()` | Es la **instanciación** (crear un objeto a partir de la clase) |







---

````python
# Create a list of messages in the familiar OpenAI format

messages = [{"role": "user", "content": "What is 2+2?"}]
````

````python
# And now call it! Any problems, head to the troubleshooting guide
# This uses GPT 4.1 nano, the incredibly cheap model
# The APIs guide (guide 9) has exact instructions for using even cheaper or free alternatives to OpenAI
# If you get a NameError, head to the guides folder (guide 6) to learn about NameErrors - always instantly fixable

response = openai_client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=messages
)

print(response.choices[0].message.content)
````


#### 🛠️ Explicación línea por línea

1. **`response = openai_client.chat.completions.create(...)`**  
   - `openai_client` → es tu **instancia del cliente** (creada con `OpenAI()` antes).  
   - `.chat.completions.create(...)` → es el **método** que envía una solicitud al modelo de chat de OpenAI.  
   - `model="gpt-4.1-nano"` → defines cuál modelo quieres usar (en este caso, uno rápido y ligero).  
   - `messages=messages` → pasas la conversación en formato de lista de mensajes, por ejemplo:
     ```python
     messages = [
         {"role": "user", "content": "What is 2+2?"}
     ]
     ```

   👉 En esta línea, básicamente le estás diciendo al modelo: *“Genera una respuesta de chat usando este modelo y estos mensajes.”*



2. **`print(response.choices[0].message.content)`**  
   - `response` → es el objeto que devuelve la API con toda la información de la respuesta.  
   - `.choices[0]` → accede a la primera opción de respuesta generada (puede haber más de una).  
   - `.message.content` → es el **texto real** que el modelo generó.  
   - `print(...)` → lo muestra en la consola.

   👉 Aquí estás imprimiendo la respuesta del modelo en pantalla.



#### 📊 Flujo resumido

| **Elemento** | **Qué hace** |
|--------------|--------------|
| **openai_client** | Tu conexión al API |
| **chat.completions.create** | Método para pedir una respuesta de chat |
| **model="gpt-4.1-nano"** | Modelo elegido |
| **messages** | Conversación enviada |
| **response.choices[0].message.content** | Texto generado por el modelo |



👉 En resumen: **primero envías los mensajes al modelo con `.create()`, y luego imprimes el contenido de la respuesta que el modelo te devuelve.**



---

````python
# And now - let's ask for a question:

question = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]
````

````python
# ask it - this uses GPT 4.1 mini, still cheap but more powerful than nano

response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

question = response.choices[0].message.content

print(question)
````

---

````python
# form a new messages list
messages = [{"role": "user", "content": question}]
````

````python
# Ask it again

response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

answer = response.choices[0].message.content
print(answer)
````
