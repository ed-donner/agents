```python
judge = f"""Estás evaluando una competición entre {len(competitors)} competidores.
A cada modelo se le ha planteado la siguiente pregunta:

{question}

Tu tarea consiste en evaluar cada respuesta en cuanto a claridad y solidez argumentativa, y clasificarlas de mejor a peor.
Responda únicamente en formato JSON, con el siguiente formato:
{{"results": ["número del mejor competidor", "número del segundo mejor competidor", "número del tercer mejor competidor", ...]}}

Estas son las respuestas de cada competidor:

{together}

Aquí están las respuestas de cada competidor:
Ahora responda con el JSON que incluya la clasificación de los competidores, sin incluir nada más. No utilice formato Markdown ni bloques de código."""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
# Judgement time!

openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
```

```python
# OK let's turn this into results!

results_dict = json.loads(results)
ranks = results_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
```

This script implements and merges two design patterns from Agile Workflows:

1. **INITIAL PATTERN: "LLM-as-a-Judge" (Tournament Phase)**
- Location: It runs just before this block.

- Explanation: A central model (gpt-5-mini) acts as a judge or static orchestrator. It receives responses from multiple LLM competitors to the same stimulus ('together') and processes an evaluation criterion to rank them in a structured ranking (JSON).

2. **ADDED/UPDATED PATTERN: "Evaluator-Optimizer" (Refinement Phase)**
- Location: It runs from the following lines using a `while` loop.

- Explanation: To take the flow to the next level, the winning model from the tournament was placed in a closed feedback loop. An evaluator analyzes its response and generates critiques.
strict, forcing the original model to correct itself and evolve
its own output in each iteration until it is approved or the allowed attempts are exhausted.

<img src="../../../assets/evaluator-optimizer.png" width="600" height="300" style="display: block;" />

```python
# =====================================================================
# IMPLEMENTACIÓN DEL PATRÓN ACTUALIZADO (EVALUATOR-OPTIMIZER)
# =====================================================================

# 1. Extracción del ganador absoluto usando el patrón LLM-as-a-Judge
best_index = int(ranks[0]) - 1
winner_model = competitors[best_index]
winner_answer = answers[best_index]

print(f"\n🚀 Iniciando fase de optimización para el ganador: {winner_model}\n")

# 2. Configuración para el patrón Evaluator-Optimizer
# ORQUESTACIÓN: Inicialización de variables para el patrón Evaluator-Optimizer
status = "RECHAZADO"                 # Estado inicial que activa el bucle
attempts = 0                         # Contador de ciclos
max_attempts = 3                     # Nuestro "escudo" para evitar bucles infinitos

current_answer = winner_answer       # Esta variable mutará en cada iteración

# 3. El Orquestador Dinámico (bucle WHILE de optimización)
while status == "RECHAZADO" and attempts < max_attempts:
    attempts += 1
    print(f"🔄 Vuelta de optimización {attempts}...")
    
    # Creamos el prompt para el Juez Evaluator-Optimizer
    optimizer_judge = f"""
    Eres un editor experto y ultra estricto. Un modelo de IA ha respondido a la siguiente pregunta:
    
    Pregunta original: {question}
    
    Esta es la respuesta actual del modelo:
    {current_answer}
    
    Tu tarea es evaluar si la respuesta es perfecta, sólida y clara, o si tiene fallas y puede mejorar.
    Responde ÚNICAMENTE en formato JSON con la siguiente estructura, sin bloques de código Markdown ni texto extra:
    {{"status": "RECHAZADO" o "APROBADO", "feedback": "Críticas detalladas de qué mejorar si fue RECHAZADO. Si fue APROBADO, déjalo vacío."}}
    """
    
    # El Evaluador genera su veredicto en formato JSON estructurado
    # Llamamos a gpt-5-mini para que evalúe esta versión
    response_eval = openai.chat.completions.create(
        model="gpt-5-mini",
        messages=[{"role": "user", "content": optimizer_judge}]
    )
    
    # El orquestador decodifica e interpreta la lógica del JSON
    # Procesamos la decisión del evaluador
    eval_result = json.loads(response_eval.choices[0].message.content)
    status = eval_result["status"]
    feedback = eval_result["feedback"]
    
    if status == "RECHAZADO":
        print(f"❌ Respuesta rechazada por el evaluador. Crítica: {feedback}\n")
        print(f"✍️ Solicitando al modelo {winner_model} que mejore su respuesta...")
        
        # El rol del Optimizador: El modelo original se corrige usando el feedback
        # Le pedimos al modelo ganador original que se corrija a sí mismo usando el feedback
        correction_prompt = f"""
        Tu respuesta anterior fue evaluada y necesita mejoras.
        
        Pregunta original: {question}
        Tu respuesta anterior: {current_answer}
        Crítica del evaluador: {feedback}
        
        Por favor, reescribe tu respuesta corrigiendo los puntos señalados por el evaluador para hacerla perfecta.
        """
        
        # Aquí usamos la lógica de tu adaptador o cliente según el tipo de modelo que haya ganado
        # Supongamos que usamos la estructura OpenAI por ahora:
        response_fix = openai.chat.completions.create(
            model=winner_model,
            messages=[{"role": "user", "content": correction_prompt}]
        )
        # Actualizamos la respuesta con la nueva versión mejorada
        current_answer = response_fix.choices[0].message.content
        
    else:
        print("✅ ¡La respuesta ha sido APROBADA por el evaluador!")

# 4. Resultado final fuera del bucle
print("\n================ CONFIGURACIÓN FINAL DEL AGENTE ================")
print(f"Modelo Ganador: {winner_model}")
print(f"Intentos requeridos: {attempts}")
print("\n📝 Respuesta final optimizada:")
display(Markdown(current_answer))
```
