### Quick tips on how to deploy the career_conversations app.py (Chatbot) on your own website/blog/portfolio to showcase

This is for those who have control of their website and can modify, build, and redeploy the index.html file.

Here is mine (Scroll at the bottom page to see where I embedded the Career Conversation component in my website):

https://imouiche.github.io/Mouichify/

Sorry! my website is still under development, but the idea is there :) 

1- The first is to host your app on Hugging Face (HF) space as taught in class/week1

2- After step1 above, you'll have the HF link to your deplyed app (e.g. https://inoussa-guru-career-conversations.hf.space).

3- We'll use this link to embed the web component into our website (index.html). This approach offers a better experience than IFrame as it does not slow down the loading time and the UI automatically adjusts its height based on the size of the Gradio app.

4- Open the index.html and create a new section/component where you'd like to add the chatbot to answer your career questions from the visitors. Below is a simple section I created for my case. Make sure to replace 5.49.2 in the Gradion S3 URL with the library version of Gradio you are using.
``` html
==============================================================
<section class="block" data-aos="zoom-in">
    <header class="block__header">
        <h2>Career Conversations</h2>
        <p>
          Feel free to ask any questions about my career in the chat.
        </p>
    </header>
        <div class="container">
            <script
        	type="module"
        	src="https://gradio.s3-us-west-2.amazonaws.com/5.34.2/gradio.js"
          ></script>
            <gradio-app src="https://inoussa-guru-career-conversations.hf.space"></gradio-app>
        </div>
</section>
================================================================================
```
https://imouiche.github.io/Mouichify/

  <img src="../../assets/career_chatbot_image.png"/>

### Read more:
- Check how to deploy apps with Gradio via the following link:

https://www.gradio.app/guides/sharing-your-app#hosting-on-hf-spaces

- Host a website on GitHub

https://www.youtube.com/watch?v=e5AwNU3Y2es


Reach out if you have any questions:

https://www.linkedin.com/in/inoussa-mouiche-b5b5138b/
