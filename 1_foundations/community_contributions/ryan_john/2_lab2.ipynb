```python
import openai  # Required to access the RateLimitError class
from openai import OpenAI

gemini = OpenAI(
    api_key=google_api_key, 
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
model_name = "gemini-2.5-flash"

def get_gemini_response(messages, retries=5, delay=2):
    import time
    for i in range(retries):
        try:
            # Your exact code block inside a protected try block
            response = gemini.chat.completions.create(
                model=model_name, 
                messages=messages
            )
            return response.choices[0].message.content
            
        except openai.RateLimitError as e:
            # If we've run out of retries, raise the error to crash safely
            if i == retries - 1:
                raise e
            
            print(f"Rate limit hit (15 RPM ceiling). Retrying in {delay} seconds...")
            time.sleep(delay)
            delay *= 2  # Exponential backoff (2s -> 4s -> 8s -> 16s)
            
        except Exception as other_error:
            # Instantly raise any non-rate-limit errors (like auth or syntax errors)
            raise other_error

answer = get_gemini_response(messages)
display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Factory Pattern Implementation for API Clients
from abc import ABC, abstractmethod

class APIClient(ABC):
    """Abstract base class for API clients"""
    
    @abstractmethod
    def create_completion(self, model_name, messages):
        """Create a completion using the specific API"""
        pass
    
    @abstractmethod
    def extract_content(self, response):
        """Extract content from the API response"""
        pass

class OpenAIClient(APIClient):
    """OpenAI API client"""
    
    def __init__(self):
        self.client = OpenAI()
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from OpenAI response"""
        return response.choices[0].message.content

class AnthropicClient(APIClient):
    """Anthropic API client"""
    
    def __init__(self):
        self.client = Anthropic()
    
    def create_completion(self, model_name, messages):
        return self.client.messages.create(
            model=model_name, 
            messages=messages, 
            max_tokens=1000
        )
    
    def extract_content(self, response):
        """Extract content from Anthropic response"""
        return response.content[0].text

class GoogleClient(APIClient):
    """Google API client"""
    
    def __init__(self):
        self.client = OpenAI(base_url='https://generativelanguage.googleapis.com/v1beta/models', api_key=google_api_key)
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from Google response"""
        return response.choices[0].message.content

class DeepSeekClient(APIClient):
    """DeepSeek API client"""
    
    def __init__(self):
        self.client = OpenAI(base_url='https://api.deepseek.com/v1', api_key=deepseek_api_key)
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from DeepSeek response"""
        return response.choices[0].message.content

class GroqClient(APIClient):
    """Groq API client"""
    
    def __init__(self):
        self.client = OpenAI(base_url='https://api.groq.com/openai/v1', api_key=groq_api_key)
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from Groq response"""
        return response.choices[0].message.content

class OllamaClient(APIClient):
    """Ollama local client"""
    
    def __init__(self):
        self.client = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from Ollama response"""
        return response.choices[0].message.content

class APIClientFactory:
    """Factory class to create appropriate API clients"""
    
    # Mapping of model names to their corresponding API clients
    MODEL_MAPPING = {
        'gpt-4o-mini': OpenAIClient,
        'o3-mini': OpenAIClient,
        'claude-3-7-sonnet-latest': AnthropicClient,
        'gemini-2.5-flash': GoogleClient,
        'deepseek-chat': DeepSeekClient,
        'llama-3.3-70b-versatile': GroqClient,
        'llama3.2': OllamaClient,
        'qwen2.5': OllamaClient,
    }
    
    @classmethod
    def create_client(cls, model_name):
        """Create the appropriate API client based on model name"""
        if model_name in cls.MODEL_MAPPING:
            client_class = cls.MODEL_MAPPING[model_name]
            return client_class()
        else:
            raise ValueError(f"Unknown model: {model_name}")
    
    @classmethod
    def get_available_models(cls):
        """Get list of available model names"""
        return list(cls.MODEL_MAPPING.keys())

print("Factory pattern implemented! Available models:")
print(APIClientFactory.get_available_models())
```

```python
# Demonstration: Using the Factory Pattern with the selected model
selected_model = model_choice_dict['model_name']
question_to_model = [{"role": "user", "content": question}]

try:
    # Create the appropriate API client using the factory
    api_client = APIClientFactory.create_client(selected_model)
    print(f"Successfully created API client for: {selected_model}")
    
    # Example usage - you can now use the client to make API calls
    response = api_client.create_completion(selected_model, question_to_model)
    answer = api_client.extract_content(response)
    # answer = response.choices[0].message.content
    display(Markdown(answer))
    
    
except ValueError as e:
    print(f"Error: {e}")
    print("Available models:", APIClientFactory.get_available_models())
```
