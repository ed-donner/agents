```python
# Factory Pattern Implementation for API Clients
from abc import ABC, abstractmethod

class APIClient(ABC):
    """Abstract base class for API clients"""
    
    @abstractmethod
    def create_completion(self, model_name, messages):
        """Create a completion using the specific API"""
        pass
    
    @abstractmethod
    def extract_content(self, response):
        """Extract content from the API response"""
        pass

class OpenAIClient(APIClient):
    """OpenAI API client"""
    
    def __init__(self):
        self.client = OpenAI()
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from OpenAI response"""
        return response.choices[0].message.content

class AnthropicClient(APIClient):
    """Anthropic API client"""
    
    def __init__(self):
        self.client = Anthropic()
    
    def create_completion(self, model_name, messages):
        return self.client.messages.create(
            model=model_name, 
            messages=messages, 
            max_tokens=1000
        )
    
    def extract_content(self, response):
        """Extract content from Anthropic response"""
        return response.content[0].text

class GoogleClient(APIClient):
    """Google API client"""
    
    def __init__(self):
        self.client = OpenAI(base_url='https://generativelanguage.googleapis.com/v1beta/models', api_key=google_api_key)
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from Google response"""
        return response.choices[0].message.content

class DeepSeekClient(APIClient):
    """DeepSeek API client"""
    
    def __init__(self):
        self.client = OpenAI(base_url='https://api.deepseek.com/v1', api_key=deepseek_api_key)
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from DeepSeek response"""
        return response.choices[0].message.content

class GroqClient(APIClient):
    """Groq API client"""
    
    def __init__(self):
        self.client = OpenAI(base_url='https://api.groq.com/openai/v1', api_key=groq_api_key)
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from Groq response"""
        return response.choices[0].message.content

class OllamaClient(APIClient):
    """Ollama local client"""
    
    def __init__(self):
        self.client = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
    
    def create_completion(self, model_name, messages):
        return self.client.chat.completions.create(
            model=model_name, 
            messages=messages
        )
    
    def extract_content(self, response):
        """Extract content from Ollama response"""
        return response.choices[0].message.content

class APIClientFactory:
    """Factory class to create appropriate API clients"""
    
    # Mapping of model names to their corresponding API clients
    MODEL_MAPPING = {
        'gpt-4o-mini': OpenAIClient,
        'o3-mini': OpenAIClient,
        'claude-3-7-sonnet-latest': AnthropicClient,
        'gemini-2.0-flash': GoogleClient,
        'deepseek-chat': DeepSeekClient,
        'llama-3.3-70b-versatile': GroqClient,
        'llama3.2': OllamaClient,
        'qwen2.5': OllamaClient,
    }
    
    @classmethod
    def create_client(cls, model_name):
        """Create the appropriate API client based on model name"""
        if model_name in cls.MODEL_MAPPING:
            client_class = cls.MODEL_MAPPING[model_name]
            return client_class()
        else:
            raise ValueError(f"Unknown model: {model_name}")
    
    @classmethod
    def get_available_models(cls):
        """Get list of available model names"""
        return list(cls.MODEL_MAPPING.keys())

print("Factory pattern implemented! Available models:")
print(APIClientFactory.get_available_models())
```

```python
# Demonstration: Using the Factory Pattern with the selected model
selected_model = model_choice_dict['model_name']
question_to_model = [{"role": "user", "content": question}]

try:
    # Create the appropriate API client using the factory
    api_client = APIClientFactory.create_client(selected_model)
    print(f"Successfully created API client for: {selected_model}")
    
    # Example usage - you can now use the client to make API calls
    response = api_client.create_completion(selected_model, question_to_model)
    answer = api_client.extract_content(response)
    # answer = response.choices[0].message.content
    display(Markdown(answer))
    
    
except ValueError as e:
    print(f"Error: {e}")
    print("Available models:", APIClientFactory.get_available_models())
```

```python
# Utility function for easy model interaction
def get_model_response(model_name, messages, display_response=True):
    """
    Utility function to get a response from any supported model
    
    Args:
        model_name (str): The name of the model to use
        messages (list): List of message dictionaries
        display_response (bool): Whether to display the response using Markdown
    
    Returns:
        str: The extracted response content
    """
    try:
        # Create the appropriate API client
        api_client = APIClientFactory.create_client(model_name)
        
        # Make the API call
        response = api_client.create_completion(model_name, messages)
        
        # Extract content using the model-specific method
        answer = api_client.extract_content(response)
        
        if display_response:
            print(f"🤖 Response from {model_name}:")
            display(Markdown(answer))
        else:
            print(f"✅ Got response from {model_name}")
            
        return answer
        
    except ValueError as e:
        print(f"❌ Error: {e}")
        print("Available models:", APIClientFactory.get_available_models())
        return None
    except Exception as e:
        print(f"❌ API Error: {e}")
        return None

# Example usage
print("🛠️ Utility Function Demo")
print("=" * 30)

# Test with a simple question
test_messages = [{"role": "user", "content": "What is 2+2?"}]
result = get_model_response(selected_model, test_messages, display_response=False)
if result:
    print(f"📊 Result: {result}")
```

```python
# Enhanced error handling for API overload and other issues
import time
import random
from functools import wraps

def retry_on_failure(max_retries=3, delay_range=(1, 3)):
    """
    Decorator to retry API calls on failure with exponential backoff
    """
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(max_retries):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    error_msg = str(e).lower()
                    
                    # Check for specific retryable errors
                    retryable_errors = [
                        'overloaded',
                        'rate limit',
                        'timeout',
                        'connection',
                        'server error',
                        'internal error',
                        'temporarily unavailable'
                    ]
                    
                    if any(error in error_msg for error in retryable_errors):
                        if attempt < max_retries - 1:
                            # Calculate delay with exponential backoff + jitter
                            delay = (delay_range[0] * (2 ** attempt)) + random.uniform(0, 1)
                            print(f"⚠️  Attempt {attempt + 1} failed: {e}")
                            print(f"🔄 Retrying in {delay:.1f} seconds...")
                            time.sleep(delay)
                            continue
                    
                    # If not retryable or max retries reached, raise the error
                    raise e
            
            return None
        return wrapper
    return decorator

# Enhanced API Client with retry logic
class RobustAPIClient:
    """Wrapper class that adds retry logic to API clients"""
    
    def __init__(self, api_client):
        self.api_client = api_client
        self.model_name = None
    
    @retry_on_failure(max_retries=3, delay_range=(1, 3))
    def create_completion_with_retry(self, model_name, messages):
        """Create completion with retry logic"""
        self.model_name = model_name
        return self.api_client.create_completion(model_name, messages)
    
    def extract_content(self, response):
        """Extract content using the underlying client's method"""
        return self.api_client.extract_content(response)

# Enhanced Factory that creates robust clients
class RobustAPIClientFactory:
    """Factory that creates API clients with retry logic"""
    
    @classmethod
    def create_robust_client(cls, model_name):
        """Create a robust API client with retry logic"""
        base_client = APIClientFactory.create_client(model_name)
        return RobustAPIClient(base_client)

print("🛡️  Enhanced error handling and retry logic implemented!")
print("📋 Available functions:")
print("  - retry_on_failure() decorator")
print("  - RobustAPIClient wrapper class")
print("  - RobustAPIClientFactory for creating robust clients")
```

```python
# Test the robust error handling with the selected model
print("🛡️  Testing robust error handling...")
print("=" * 40)

selected_model = model_choice_dict['model_name']
test_message = [{"role": "user", "content": "Hello! Please respond briefly."}]

try:
    # Create a robust client that handles overload errors
    robust_client = RobustAPIClientFactory.create_robust_client(selected_model)
    print(f"✅ Created robust client for: {selected_model}")
    
    # Make API call with automatic retry on overload
    print(f"🔄 Making API call (will retry on overload)...")
    response = robust_client.create_completion_with_retry(selected_model, test_message)
    
    # Extract content
    answer = robust_client.extract_content(response)
    print(f"✅ Success! Response: {answer}")
    
except Exception as e:
    print(f"❌ Final error after all retries: {e}")
    print("💡 This might be a persistent issue - try a different model or wait a bit")

# Alternative: Try a different model if the first one fails
print(f"\n🔄 Alternative: Trying with a backup model...")
backup_models = ['gpt-4o-mini', 'llama3.2', 'qwen2.5']

for backup_model in backup_models:
    if backup_model != selected_model:
        try:
            print(f"🧪 Testing {backup_model}...")
            backup_client = RobustAPIClientFactory.create_robust_client(backup_model)
            response = backup_client.create_completion_with_retry(backup_model, test_message)
            answer = backup_client.extract_content(response)
            print(f"✅ {backup_model} worked! Response: {answer}")
            break
        except Exception as e:
            print(f"❌ {backup_model} also failed: {e}")
            continue
```
