```python
# First let's do an import. If you get an Import Error, see Q5 and then Q25 here: https://edwarddonner.com/avatar

from dotenv import load_dotenv
from openai import OpenAI

openai = OpenAI()
```

```python
# Next it's time to load the API keys into environment variables
# If this returns false, see the next cell!

load_dotenv(override=True)
```

```python
from IPython.display import Markdown, display

# First create the messages:

prompt = "Please propose a business area that is worth exploring for Agentic AI product opportunities."
messages = [{"role": "user", "content": prompt}]

# Then make the first call:

response = openai.chat.completions.create(model="gpt-5.4", messages=messages)

# Then read the business idea:

businessArea = response.choices[0].message.content
display(Markdown(businessArea))

# Ask for the challenges in the areas that could need an agentic AI solution:

prompt = "Please provide a pain-point in that business industry; something challenging that may be a great opportunity for an Agentic solution."
messages = [{"role": "user", "content": prompt}]

response = openai.chat.completions.create(model="gpt-5.4", messages=messages)

# read the area challenges:

challenge = response.choices[0].message.content
display(Markdown(challenge))

# Ask for an Agentic AI solution for the challenge:

prompt = "Please provide an Agentic AI solution for the challenge."
messages = [{"role": "user", "content": prompt}]

response = openai.chat.completions.create(model="gpt-5.4", messages=messages)

# read the challenge solution:

solution = response.choices[0].message.content
display(Markdown(solution))

# And repeat! In the next message, include the business area within the message
```
