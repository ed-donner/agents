```python
import os

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")

if GEMINI_API_KEY: 
    print(f"GEMINI_API_KEY is set and begins with {GEMINI_API_KEY[:8]}")
else:
    print("GEMINI_API_KEY is not set")
```

```python
messages = [
    {
        "role": "user",
        "parts": [{"text": "Hello, I have a question about my code."}]
    },
    {
        "role": "model",
        "parts": [{"text": "Sure, I can help with that! What is the issue?"}]
    },
    {
        "role": "user",
        "parts": [{"text": "pick a business area that might be worth exploring for an Agentic AI opportunity?"}] # Added this
    }
]
```

```python
import os
from google import genai
api_key = os.environ.get("GEMINI_API_KEY")
# print(api_key)
client = genai.Client(api_key=api_key)

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=messages,
)

question = response.text+"present a pain-point in that industry - something challenging that might be ripe for an Agentic solution"
print(question)
```
